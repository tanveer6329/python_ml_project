```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_coal_data = pd.read_csv('coalv2.csv')
raw_coal_data.head()
```

```python
coal = raw_coal_data.iloc[:5000, ]
coal = coal.round(1)
```

```python
#plt.style.use('ggplot')

fig, axs = plt.subplots(3, 3, sharey=True)
fig.set_figheight(8)
fig.set_figwidth(8)


axs[0, 0].scatter(coal['moistr'], coal['gcv'])
axs[0, 0].set_xticklabels(coal['moistr'])
axs[0, 0].set_yticklabels(coal['gcv'])
axs[0, 0].set_xlabel('% Moisture ("As received" basis)', fontsize=12)
axs[0, 0].set_ylabel('GCV (BTU/lb)', fontsize=12)

axs[0, 1].scatter(coal['volmat'], coal['gcv'])
axs[0, 1].set_xticklabels(coal['volmat'], weight='bold')
#axs[0, 1].set_yticklabels(axs[0, 1].get_yticks(), weight='bold')
axs[0, 1].set_xlabel('% Volatile matter ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[0, 1].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

axs[0, 2].scatter(coal['fixedc'], coal['gcv'])
axs[0, 2].set_xticklabels(coal['fixedc'], weight='bold')
#axs[0, 2].set_yticklabels(axs[0, 2].get_yticks(), weight='bold')
axs[0, 2].set_xlabel('% Fixed carbon ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[0, 2].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

axs[1, 0].scatter(coal['ash'], coal['gcv'])
axs[1, 0].set_xticklabels(coal['ash'], weight='bold')
axs[1, 0].set_yticklabels(coal['gcv'], weight='bold')
axs[1, 0].set_xlabel('% Ash ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
axs[1, 0].set_ylabel('GCV (BTU/lb)', fontsize=17, fontweight='bold', fontname='Times New Roman')

axs[1, 1].scatter(coal['hydrgn'], coal['gcv'])
axs[1, 1].set_xticklabels(coal['hydrgn'], weight='bold')
#axs[1, 1].set_yticklabels(axs[1, 1].get_yticks(), weight='bold')
axs[1, 1].set_xlabel('% H ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[1, 1].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

axs[1, 2].scatter(coal['carbon'], coal['gcv'])
axs[1, 2].set_xticklabels(coal['carbon'], weight='bold')
#axs[1, 2].set_yticklabels(axs[1, 2].get_yticks(), weight='bold')
axs[1, 2].set_xlabel('% C ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[1, 2].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

axs[2, 0].scatter(coal['nitrgn'], coal['gcv'])
axs[2, 0].set_xticklabels(coal['nitrgn'], weight='bold')
axs[2, 0].set_yticklabels(coal['gcv'], weight='bold')
axs[2, 0].set_xlabel('% N ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
axs[2, 0].set_ylabel('GCV (BTU/lb)', fontsize=17, fontweight='bold', fontname='Times New Roman')

axs[2, 1].scatter(coal['oxygen'], coal['gcv'])
axs[2, 1].set_xticklabels(coal['oxygen'], weight='bold')
#axs[2, 1].set_yticklabels(axs[2, 1].get_yticks(), weight='bold')
axs[2, 1].set_xlabel('% O ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[2, 1].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

axs[2, 2].scatter(coal['sulfur'], coal['gcv'])
axs[2, 2].set_xticklabels(coal['sulfur'], weight='bold')
#axs[2, 2].set_yticklabels(axs[2, 2].get_yticks(), weight='bold')
axs[2, 2].set_xlabel('% S ("As received" basis)', fontsize=17, fontweight='bold', fontname='Times New Roman')
#axs[2, 2].set_ylabel('GCV (BTU/lb)', fontsize=15, fontweight='bold', fontname='Times New Roman')

fig.tight_layout()
plt.savefig('cross_plot11.png', dpi=400)
```
