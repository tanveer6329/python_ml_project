```python
import struct
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('coalv2.csv')
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
x.head()
```

```python
y.head()
```

```python
# taking every column values and storing it into a python list
moistr =[v for v in x['moistr']]
volmat =[v for v in x['volmat']]
fixedc =[v for v in x['fixedc']]
ash =[v for v in x['ash']]
hydrgn =[v for v in x['hydrgn']]
carbon =[v for v in x['carbon']]
nitrgn =[v for v in x['nitrgn']]
oxygen =[v for v in x['oxygen']]
sulfur =[v for v in x['sulfur']]
```

```python
moistr_b = []
volmat_b = []
fixedc_b = []
ash_b = []
hydrgn_b = []
carbon_b = []
nitrgn_b = []
oxygen_b = []
sulfur_b = []
```

```python
# function for converting floating point value to 32 bit binary
def float_to_bin(num):
    return format(struct.unpack('!I', struct.pack('!f', num))[0], '032b')

# code for converting float to 32-bit binary and adding in new list
for num in moistr:
  moistr_b.append(float_to_bin(num))


for num in volmat:
  volmat_b.append(float_to_bin(num))


for num in fixedc:
  fixedc_b.append(float_to_bin(num))


for num in ash:
  ash_b.append(float_to_bin(num))


for num in hydrgn:
  hydrgn_b.append(float_to_bin(num))


for num in carbon:
  carbon_b.append(float_to_bin(num))


for num in nitrgn:
  nitrgn_b.append(float_to_bin(num))


for num in oxygen:
  oxygen_b.append(float_to_bin(num))


for num in sulfur:
  sulfur_b.append(float_to_bin(num))
```

```python
# creating a dataframe with all the binary values
df2 = pd.DataFrame(zip(moistr_b, volmat_b, fixedc_b, ash_b, hydrgn_b, carbon_b, nitrgn_b, oxygen_b, sulfur_b), 
               columns =['moistr', 'volmat', 'fixedc', 'ash', 'hydrgn', 'carbon', 'nitrgn',
       'oxygen', 'sulfur']) 
print(df2.shape)
df2.head(10)
```

```python
# transposing the dataframe
df3 = df2.transpose()
df3.head(10)
```

```python
df4 = df3[0]
df4.to_csv('binaryimage.csv')
```

```python
# creating an empty list and appending numbers of empty list within it
l = []
for i in range(6582):
	l.append([])

print(len(l))
l[0:10]
```

```python
for x in df3.columns:
	l[x] = [y for y in df3[x]]

l[0]
```

```python
data = []
for i in range(len(l)):
    data.append([])
    
print(data[0:20])
len(data)
```

```python
n=0
for list in l: # entering into nested list
    for string in list: # entering into string
        for value in string: # entering into value of the string
            data[n].append(value)
    n=n+1
```

```python
print(data[0])
```

```python
#converting every 0/1 string value to integer
for list in data:
    for i in range(0, len(list)): 
        list[i] = int(list[i]) 
```

```python
print(data[0])
len(data[0])
```

```python
# converting 1 to 255
for i in range(len(data)):
	for j in range(len(data[i])):
		data[i][j] = data[i][j]*255
print(data[0])
```
